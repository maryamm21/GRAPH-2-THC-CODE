```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for the simplified graph
data = {
    'Cannabis Use Group': [
        'Never Used Cannabis (Baseline)',
        'Daily High-Potency Cannabis Use'
    ],
    'Mean Age of Onset (Years)': [31.4, 25.2]
}
df = pd.DataFrame(data)

# Create the bar chart
plt.figure(figsize=(7, 6))

# Define colors 
colors = ['#1E5631', '#4CAF50'] 

bars = plt.bar(df['Cannabis Use Group'], df['Mean Age of Onset (Years)'], color=colors)

# Add labels and title
plt.title('Mean Age of Onset of Psychosis (AOP):\nNever vs. Daily High-Potency Use', fontsize=14)
plt.xlabel('Cannabis Use Group', fontsize=12)
plt.ylabel('Mean Age of Onset (Years)', fontsize=12)

# Set the y-axis limit to focus on the difference in onset
plt.ylim(0, 33)

# Add data labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f'{yval:.1f}', ha='center', va='bottom', fontsize=11, fontweight='bold')

# Calculate the difference for annotation
difference = df['Mean Age of Onset (Years)'].iloc[0] - df['Mean Age of Onset (Years)'].iloc[1]

# Ensure layout is tight to prevent label truncation
plt.tight_layout()

# Save the figure
plt.savefig('simplified_aop_thc.png')
```
